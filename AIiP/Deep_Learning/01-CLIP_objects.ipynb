```python
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import torch, os

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(DEVICE)
proc  = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
#dodanie dodatkowych labeli i powiekszenie datasetu
labels  = ["cat", "dog", "drone", "car", "ship", "cow", "sheep"]
prompts = [f"a photo of a {l}" for l in labels]
#sprawdzenie wynikow

files = [f"images/{f}" for f in os.listdir("images") if f.lower().endswith((".jpg", ".png", ".jpeg"))][:17]

for fp in files:
    img = Image.open(fp).convert("RGB")
    inputs = proc(text=prompts, images=img, return_tensors="pt", padding=True).to(DEVICE)
    with torch.inference_mode():
        logits = model(**inputs).logits_per_image[0]
        probs  = logits.softmax(-1)
    pred = labels[int(probs.argmax())]
    print(os.path.basename(fp), "->", pred)
```
