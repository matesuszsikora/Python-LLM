```python
from ultralytics import YOLO
model = YOLO("yolo11n.pt")
model.predict(source="detect/images/val", imgsz=640, conf=0.25, save=True)
model.train(data="detect/data.yaml", epochs=3, imgsz=640, batch=16)
model.val()
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

model = YOLO("runs/detect/train/weights/best.pt")  # detection weights
results = model.predict(source="detect/images/val", imgsz=640, conf=0.25, save=True)

# Ultralytics auto-saves images with boxes in runs/detect/predict...
# Show them inline too:
for r in results:
    im_bgr = r.plot()   # draws boxes + labels
    im_rgb = im_bgr[:, :, ::-1]
    plt.figure(figsize=(8,6)); plt.imshow(im_rgb); plt.axis("off"); plt.show()
```
